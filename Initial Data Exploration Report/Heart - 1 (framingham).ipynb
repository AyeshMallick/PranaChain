```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
file_path = r"C:\Users\Sandeep\Desktop\MAC\3 Sem\Internship\HEART 1 - framingham.csv"
df = pd.read_csv(file_path)
```

```python
# ----- 1. Basic Dataset Information -----
print("\n--- Dataset Info ---")
print(df.info())

print("\n--- First 5 Rows ---")
print(df.head())

print("\n--- Summary Statistics ---")
print(df.describe(include="all"))
```

```python
# ----- 2. Column Explanations -----
column_explanations = {
    "male": "Gender of participant (0 = Female, 1 = Male)",
    "age": "Age in years",
    "education": "Education level (1–4)",
    "currentSmoker": "Whether the participant is a current smoker (0 = No, 1 = Yes)",
    "cigsPerDay": "Number of cigarettes smoked per day",
    "BPMeds": "Whether participant is on blood pressure medication (0 = No, 1 = Yes)",
    "prevalentStroke": "History of stroke (0 = No, 1 = Yes)",
    "prevalentHyp": "History of hypertension (0 = No, 1 = Yes)",
    "diabetes": "Diabetes diagnosis (0 = No, 1 = Yes)",
    "totChol": "Total cholesterol level (mg/dL)",
    "sysBP": "Systolic blood pressure (mmHg)",
    "diaBP": "Diastolic blood pressure (mmHg)",
    "BMI": "Body Mass Index (kg/m²)",
    "heartRate": "Resting heart rate (beats per minute)",
    "glucose": "Fasting blood glucose (mg/dL)",
    "TenYearCHD": "Target variable: 10-year risk of coronary heart disease (0 = No, 1 = Yes)"
}

print("\n--- Column Explanations ---")
for col, desc in column_explanations.items():
    print(f"{col}: {desc}")

# ----- 3. Check Missing Values -----
print("\n--- Missing Values ---")
print(df.isnull().sum())

# ----- 4. Correlation Analysis -----
corr = df.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", annot=False, cbar=True)
plt.title("Correlation Heatmap of Features")
plt.show()

# Correlation of each feature with target (TenYearCHD)
target_corr = corr["TenYearCHD"].sort_values(ascending=False)
print("\n--- Correlation with TenYearCHD ---")
print(target_corr)

# ----- 5. Pairwise Plots for Key Features -----
key_features = ["age", "cigsPerDay", "totChol", "sysBP", "BMI", "glucose", "TenYearCHD"]
sns.pairplot(df[key_features], hue="TenYearCHD")
plt.show()
```
